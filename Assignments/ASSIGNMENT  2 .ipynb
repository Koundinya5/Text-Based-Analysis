# Regex 

```python
import regex as re# imports the regex

def tokenize(text):
    return re.findall(r'[\w-]*[a-z|A-Z][\w-]*', text)
```

```python
testtext = "Burn it to the ashes mate# www.iftar.asylum"
testtokens = tokenize(testtext)# tokenizes the given sample
 
print(testtokens)
```

# Stopwords

```python
import nltk 
nltk.download('stopwords')
stopwords = set(nltk.corpus.stopwords.words('english'))
print(stopwords)
# imports the inbuilt stopwords depending upon the language specified
```

```python
def remove_stop(tokens):
    return [t for t in tokens if t.lower() not in stopwords]
#Just testing it
cleantokens= remove_stop(testtokens)
print(cleantokens)
#removing the stopwords
```

# Intialization the pipeline

```python
pipeline = [str.lower, tokenize, remove_stop]

def prepare(text, pipeline):
    tokens = text
    for fun in pipeline:
        tokens = fun(tokens)
    return tokens
```

```python
df['tokens'] = df['artist_verses'].apply(prepare, pipeline=pipeline)# runs the pipeline
```

# Counting the tokens in the dataset

```python
df['num_tokens'] = df['tokens'].map(len)
df.tail()
```

# Word Frequency Analysis

# Using the Counter

```python
from collections import Counter

tokens = tokenize("The way this way looks reminds me of the way we took to find this way.")

testcount= Counter(tokens)
print(testcount)
# testing the counter
```

```python
more_tokens = tokenize("She likes dogs and cats.")# adding new data for the already existing count
testcount.update(more_tokens)
print(testcount)
```

```python
counter = Counter()
df['tokens'].map(counter.update)# using counter on our dataset
print(counter.most_common(5))
```

```python
freq_df = pd.DataFrame.from_dict(counter, orient='index', columns=['freq'])
freq_df = freq_df.query('freq >= 1')# specifying the frequency of the word count
freq_df.index.name = 'token'
```

```python
freq_df = freq_df.sort_values('freq', ascending=False)
freq_df.head(20)
# sorting the data and displaying it
```

```python
freq_df.describe()# stats
```
