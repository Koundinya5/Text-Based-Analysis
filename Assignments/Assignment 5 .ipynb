# Sentiment analysis

# Sentiment analysis is the use of natural language processing to gauge the emotional tone in text, categorizing it as positive, negative, or neutral. This helps understand public opinion, customer feedback, and social media trends.

```python
import pandas as pd 
df = pd.read_csv( r'C:\Users\ASUS\Downloads\lyrics_raw.csv')
df.sample(10)
```

```python
from sklearn import preprocessing
import nltk
nltk.download('opinion_lexicon')
from nltk.corpus import opinion_lexicon
from nltk.tokenize import word_tokenize

print('Total number of words in opinion lexicon', len(opinion_lexicon.words()))
print('Examples of positive words in opinion lexicon',
      opinion_lexicon.positive()[:10])
print('Examples of negative words in opinion lexicon',
      opinion_lexicon.negative()[:10])
```

```python
# Let's create a dictionary which we can use for scoring our track_name text
nltk.download('punkt')
df.rename(columns={"track_name": "text"}, inplace=True)
pos_score = 1
neg_score = -1
word_dict = {}
 
# Adding the positive words to the dictionary
for word in opinion_lexicon.positive():
        word_dict[word] = pos_score
      
# Adding the negative words to the dictionary
for word in opinion_lexicon.negative():
        word_dict[word] = neg_score
```

# The bing_liu_score function calculates the sentiment score of a given text based on a pre-defined dictionary of words and their associated sentiment scores. It tokenizes the input text into words, converts them to lowercase, and iterates through each word to check if it exists in the pre-defined dictionary. If a word is found in the dictionary, its sentiment score is added to the overall sentiment score of the text. Finally, the function returns the total sentiment score of the text.

```python
def bing_liu_score(text):
    sentiment_score = 0
    bag_of_words = word_tokenize(text.lower())
    for word in bag_of_words:
        if word in word_dict:
            sentiment_score += word_dict[word]
    return sentiment_score
```

# Calculating the Bing_Liu_Score

# This code snippet fills missing values in the 'text' column of a DataFrame named df with the string 'no track'. Then, it applies the bing_liu_score function to each text in the 'text' column and stores the results in a new column named 'Bing_Liu_Score'. This process enables sentiment scoring of the text data in the DataFrame.

```python
# Fill missing values in 'text' column with 'no track'
df['text'].fillna('no track', inplace=True)

# Apply bing_liu_score function to each text in 'text' column and store results in 'Bing_Liu_Score' column
df['Bing_Liu_Score'] = df['text'].apply(bing_liu_score)
```

# This code groups the DataFrame df by the 'artist' column and calculates the mean of the 'Bing_Liu_Score' for each group. This allows for analyzing the average sentiment score of the texts associated with each artist in the DataFrame

```python
df.groupby('artist').agg({'Bing_Liu_Score':'mean'})
```
